```
import numpy as np
import math
from random import seed
from random import randint

# تعریف متغیر ها
global tedad_fram
global tedad_dorbin
global mokhtasat_vaghei_noghat_1
global mokhtasat_vaghei_noghat_2
global mokhtasat_tasvir_noghat_1
global mokhtasat_tasvir_noghat_2
global parametr_dakheli
global parametr_kharegi
global arze_tasvir
global ertefa_tasvir
global sin_zaviye
global cos_zaviye
global zaviye
global bordar_enteghal
global zarib_evegag_k1
global zarib_evegag_k2
global zarib_evegag_p1
global zarib_evegag_p2
global parametr_dakheli
file_name_img1= "mokhtasatvaghei1.txt"
file_name_img2= "mokhtasatvaghei2.txt"
file_name_img_1= "camera1.txt"
file_name_img_2= "camera2.txt"
```

```

def meghdardehi_avaliye():
    global tedad_fram
    global tedad_dorbin
    global parametr_dakheli
    global parametr_kharegi
    global arze_tasvir
    global ertefa_tasvir
    global zaviye
    global bordar_enteghal
    global zarib_evegag_k1
    global zarib_evegag_k2
    global zarib_evegag_p1
    global zarib_evegag_p2

    tedad_fram=50
    tedad_dorbin=3
    parametr_dakheli=[]
    parametr_kharegi=[]
    arze_tasvir=720
    ertefa_tasvir=586
    zaviye=[[0,0,0],[-0.0214,0.0468,0],[-0.0681,0.0915,0]]
    bordar_enteghal=[[0,0,0],[14.432,1.464,0],[34.548,1.747,0]]
    zarib_evegag_k1=[-0.63950,-0.54183,-0.52066]
    zarib_evegag_k2=[0.71665,0.91538,0.82959]
    zarib_evegag_p1=[0.00183,0.00135,0.00188]
    zarib_evegag_p2=[-0.00814,0.00775,0.00734]
    parametr_dakheli=[]
    parametr_dakheli.append(np.array([[1747.9,0,369.56],[0,1781.1,280.59],[0,0,1]]))
    parametr_dakheli.append(np.array([[1639.3,0,359.65],[0,1687.1,287.87],[0,0,1]]))
    parametr_dakheli.append(np.array([[1789.5,0,367.29],[0,1798.8,278.43],[0,0,1]]))

```

```
def eijad_noghat_tasadofi():
    #ایجاد نقاط تصادفی در مختصات جهانی  با فاصله 50
    global mokhtasat_vaghei_noghat_1
    global mokhtasat_vaghei_noghat_2
    mokhtasat_vaghei_noghat_1=[]
    mokhtasat_vaghei_noghat_2=[]
    for i in range(tedad_fram):
        phi=math.radians(randint(0, 360))
        teta=math.radians(randint(0, 360))
        X_1=(randint(50, arze_tasvir-50))
        Y_1=(randint(50, ertefa_tasvir-50))
        Z_1=(randint(2000,3000))
        X_2=50*math.sin(phi)*math.cos(teta)+X_1
        Y_2=50*math.sin(phi)*math.sin(teta)+Y_1
        Z_2=50*math.cos(phi)+Z_1
        mokhtasat_vaghei_noghat_1.append(np.array([X_1,Y_1,Z_1,1]))
        mokhtasat_vaghei_noghat_2.append(np.array([X_2,Y_2,Z_2,1]))
```

```
def mohasebe_parametr_kharegi(tedad_dorbin,bordar_enteghal,zaviye):
    global parametr_kharegi
    sin_zaviye=[]
    cos_zaviye=[]
    #محاسبه نسبت های مثلثاتی مورد نیاز برای محاسبه ماتریس پارامتر های خارجی
    for i in range(len(zaviye)):
        for j in range(len(zaviye[i])):
            sin_zaviye.append(math.sin(zaviye[i][j]))
            cos_zaviye.append(math.cos(zaviye[i][j]))
    # محاسبه ماتریس پارامترهای خارجی
    for i in range(tedad_dorbin):
        zarf=[]
        zarf2=[]
        for j in range (12):
            if j==0:
                zarf.append(cos_zaviye[i*3+1]*cos_zaviye[i*3+2])
            elif j==1:
                zarf.append(-cos_zaviye[i*3+1]*sin_zaviye[i*3+2])
            elif j==2:
                zarf.append(sin_zaviye[i*3+1])
            elif j==4:
                zarf.append(cos_zaviye[i*3+0]*sin_zaviye[i*3+2]+cos_zaviye[i*3+2]*sin_zaviye[i*3+0]*sin_zaviye[i*3+1])
            elif j==5:
                zarf.append(cos_zaviye[i*3+0]*cos_zaviye[i*3+2]-sin_zaviye[i*3+0]*sin_zaviye[i*3+2]*sin_zaviye[i*3+1])
            elif j==6:
                zarf.append(-cos_zaviye[i*3+1]*sin_zaviye[i*3+0])
            elif j==8:
                zarf.append(sin_zaviye[i*3+0]*sin_zaviye[i*3+2]-cos_zaviye[i*3+2]*cos_zaviye[i*3+0]*sin_zaviye[i*3+1])
            elif j==9:
                zarf.append(cos_zaviye[i*3+2]*sin_zaviye[i*3+0]+cos_zaviye[i*3+0]*sin_zaviye[i*3+2]*sin_zaviye[i*3+1])
            elif j==10:
                zarf.append(cos_zaviye[i*3+0]*cos_zaviye[i*3+1])
            else:
                if j==3 or j==7 or j==11 :
                    zarf.append(bordar_enteghal[i][j%3])

                    zarf2.append(np.array(zarf))
                    zarf=[]
        parametr_kharegi.append(np.array(zarf2))
        zarf2=[]

```

```
def mohasebe_mokhtasat_tasvir_noghat(tedad_fram,tedad_dorbin,parametr_dakheli,parametr_kharegi,zarib_evegag_k1,zarib_evegag_k2,zarib_evegag_p1,zarib_evegag_p2,mokhtasat_vaghei_noghat_1,mokhtasat_vaghei_noghat_2,arze_tasvir,ertefa_tasvir):
    global mokhtasat_tasvir_noghat_1
    global mokhtasat_tasvir_noghat_2
    mokhtasat_tasvir_noghat_1=[]
    mokhtasat_tasvir_noghat_2=[]
    for i in range(tedad_dorbin):
        zarf=[]
        zarf1=[]
        for j in range(tedad_fram):
            zarf2=[]
            for k in range(2):
                if k==0:
                    zarf2.append(parametr_kharegi[i].dot(mokhtasat_vaghei_noghat_1[j]))
                else:
                    zarf2.append(parametr_kharegi[i].dot(mokhtasat_vaghei_noghat_2[j]))
                x_prin=zarf2[k][0]/zarf2[k][2]
                y_prin=zarf2[k][1]/zarf2[k][2]
                R_2=(x_prin**2)+(y_prin**2)
                chand_jomlei=(1+zarib_evegag_k1[i]*R_2+zarib_evegag_k2[i]*(R_2**2))
                x_zegond=x_prin*chand_jomlei+2*zarib_evegag_p1[i]*x_prin*y_prin+zarib_evegag_p2[i]*(R_2+2*(x_prin**2))
                y_zegond=y_prin*chand_jomlei+zarib_evegag_p1[i]*(R_2+2*(y_prin**2))+2*zarib_evegag_p2[i]*x_prin*y_prin
                zarf3=np.array([x_zegond,y_zegond,1])
                asghar_kolar=zarf3
                zarf3=parametr_dakheli[i].dot(zarf3)
                if k==0:
                   # if min(zarf3)>0 and zarf3[0]<arze_tasvir and zarf3[1]<ertefa_tasvir:
                        zarf.append([zarf3[0],zarf3[1]])
              #      else:
              #          zarf.append(['N','N'])
                else:
                   # if min(zarf3)>0 and zarf3[0]<arze_tasvir and zarf3[1]<ertefa_tasvir:
                        zarf1.append([zarf3[0],zarf3[1]])
                   # else:
                   #     zarf1.append(['N','N'])
        mokhtasat_tasvir_noghat_1.append(zarf)
        mokhtasat_tasvir_noghat_2.append(zarf1)
    mokhtasat_tasvir_noghat_1=np.array(mokhtasat_tasvir_noghat_1)
    mokhtasat_tasvir_noghat_2=np.array(mokhtasat_tasvir_noghat_2)
```

```

meghdardehi_avaliye()
eijad_noghat_tasadofi()
mohasebe_parametr_kharegi(tedad_dorbin,bordar_enteghal,zaviye)
mohasebe_mokhtasat_tasvir_noghat(tedad_fram,tedad_dorbin,parametr_dakheli,parametr_kharegi,zarib_evegag_k1,zarib_evegag_k2,zarib_evegag_p1,zarib_evegag_p2,mokhtasat_vaghei_noghat_1,mokhtasat_vaghei_noghat_2,arze_tasvir,ertefa_tasvir)

```
